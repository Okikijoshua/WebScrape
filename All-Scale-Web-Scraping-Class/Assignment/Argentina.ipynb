```python
import requests
from bs4 import BeautifulSoup

url = 'https://en.wikipedia.org/wiki/Argentina'
responce = requests.get(url)
soup = BeautifulSoup(responce.content, 'html.parser')

soup
```

```python
title =soup.title.text
title
```

```python
body = soup.body.text
body
```

```python
ipt = soup.script.text
ipt
```

```python
script = soup.script.contents
script
```

```python
title = soup.find('title').text
title
```

```python
table_view = soup.find('table', class_ ='infobox ib-country vcard')
table_view
```

```python
table_view = soup.find('figure', class_ ='mw-default-size mw-halign-left')
table_view
```

```python
table_view = soup.find_all('figure', class_ ='mw-default-size mw-halign-left')
for i in table_view:
    getlink = i.find('a',href=True)
    print(getlink['href'])
```
